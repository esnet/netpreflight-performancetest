```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
from pylab import *
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 40})
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
rc('axes', linewidth=4)
plt.rcParams["figure.figsize"] = (21, 15)
```

```python
labels = ['128', '512', '1024', '2048', '3072']

iperf = [133, 216, 250, 441, 751]
preflight = [160.1509901, 300.5565323, 359.7394409, 580.4443962, 785.2579103]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, iperf, width, label='iperf')
rects2 = ax.bar(x + width/2, preflight, width, label='preflight')

# Add some text for labels, title and custom x-axis tick labels, etc.


ax.set_ylabel('Throughput Mean')
ax.set_xlabel('Buffer Size (S) in (KB)')
ax.set_title('5MB')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
fig.tight_layout()
plt.show()
```
